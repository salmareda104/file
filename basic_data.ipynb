```python
import pandas as pd

# اقرأ ملف الـ Excel
joblist = pd.read_excel(r'C:\Users\salma.elsayed\Desktop\basic data\rah\joblist.xlsx')

# تنظيف البيانات من المسافات الزائدة (للتأكد من تطابق الأسماء بشكل دقيق)
joblist['Position - Job List'] = joblist['Position - Job List'].str.strip()
joblist['Function'] = joblist['Function'].str.strip()
joblist['Department'] = joblist['Department'].str.strip()

# Function to apply the condition
def adjust_position(row, grouped_df):
    position_parts = row["Position - Job List"].split(" - ")  # Split Position into parts
    base_position = position_parts[0].strip().lower()  # Get the base position (before the '-')
    
    # Get all positions in the same company group
    group = grouped_df.get_group(row['Company'])
    
    # Check if all rows in the group have the same Function and Department
    if all((g_row["Function"].strip().lower() == row["Function"].strip().lower() and g_row["Department"].strip().lower() == row["Department"].strip().lower()) for _, g_row in group.iterrows()):
        # Get all base positions (before '-')
        base_positions = [pos.split(' - ')[0].strip().lower() for pos in group["Position - Job List"]]
        
        # Check if all base positions are the same (or very similar)
        if len(set(base_positions)) == 1:  # All base titles are the same
            return base_position  # Return the base position only (before '-')
        else:
            # Check if positions after the '-' are the same but Function/Department differ
            position_after_split = [pos.split(" - ")[1].strip().lower() for pos in group["Position - Job List"]]
            if len(set(position_after_split)) == 1:  # If positions after the '-' are similar but Function/Department differ
                return row["Position - Job List"]  # Keep original title
            else:
                return row["Position - Job List"]  # Keep original if base positions differ
    else:
        return row["Position - Job List"]  # Keep original if Function/Department differ

# Group by Company
grouped = joblist.groupby('Company')

# Apply the adjustment logic
joblist["Adjusted Position"] = joblist.apply(lambda row: adjust_position(row, grouped), axis=1)

joblist[joblist['Position - Job List']=='bim modeler engineer - arch']
```

```python
from rapidfuzz import process, fuzz
import pandas as pd
import re
joblist = pd.read_excel(r'C:\Users\salma.elsayed\Desktop\basic data\rah\joblist.xlsx')

# وظيفة لتحديد مستوى الوظيفة (Hierarchy)
def extract_hierarchy_level(position):
    hierarchy_levels = [
        'junior', 
        'senior', 
        'manager', 
        'section head', 
        'director', 
        'chief', 
        'lead', 
        'team leader',
        'specialist',
        'section head',
        'senior manager',
        'Manager',
        'Supervisor'

    ]
    for level in hierarchy_levels:
        if level in position.lower():
            return level
    return "general"  # المستوى الافتراضي إذا لم يكن هناك مستوى محدد

# وظيفة لفحص التشابه مع السماح بالأخطاء الإملائية
def match_exact_position_with_tolerance(joblist_position, sap_positions, threshold=85):
    matches = process.extract(joblist_position, sap_positions, scorer=fuzz.ratio, limit=1)
    if matches and matches[0][1] >= threshold:
        return matches[0][0]  # Return the closest match if similarity meets the threshold
    return None

# وظيفة لتحديد الوظائف المتطابقة بنفس الشركة وبأخطاء بسيطة فقط مع التحقق من المستوى الوظيفي
def find_matching_positions(joblist_df, sap_df, threshold=90):
    matched_positions = []  # قائمة الوظائف المطابقة
    unmatched_positions = []  # قائمة الوظائف غير المطابقة

    for _, job in joblist_df.iterrows():
        # فلترة بيانات SAP لنفس الشركة
        sap_filtered = sap_df[sap_df['Company'] == job['Company']]
        
        # الحصول على الوظائف في SAP لنفس الشركة
        sap_positions = sap_filtered['Position'].tolist()
        
        # البحث عن التطابق
        matched_position = match_exact_position_with_tolerance(job['Position - Job List'], sap_positions, threshold)
        
        if matched_position:
            # استخراج مستوى الوظيفة من Joblist و SAP
            job_hierarchy = extract_hierarchy_level(job['Position - Job List'])
            matched_hierarchy = extract_hierarchy_level(matched_position)
            
            # التحقق من تطابق المستوى الوظيفي
            if job_hierarchy == matched_hierarchy or job_hierarchy == "general" or matched_hierarchy == "general":
                matched_positions.append({
                    'Company': job['Company'],
                    'Joblist Position': job['Position - Job List'],
                    'SAP Position': job['Position - Job List'],  # تطابق الوظيفة مع Joblist Position
                    'SAP Position (Original)': matched_position,  # النسخة الأصلية
                    'Hierarchy': job_hierarchy,
                    'Status': 'Matched'
                })
            else:
                unmatched_positions.append({
                    'Company': job['Company'],
                    'Joblist Position': job['Position - Job List'],
                    'SAP Position': matched_position,
                    'SAP Position (Original)': matched_position,  # النسخة الأصلية
                    'Hierarchy': f"Joblist: {job_hierarchy}, SAP: {matched_hierarchy}",
                    'Status': 'Hierarchy Mismatch'
                })
        else:
            unmatched_positions.append({
                'Company': job['Company'],
                'Joblist Position': job['Position - Job List'],
                'SAP Position': None,
                'SAP Position (Original)': None,  # لا توجد نسخة أصلية
                'Hierarchy': None,
                'Status': 'Not Matched'
            })
    
    return pd.DataFrame(matched_positions), pd.DataFrame(unmatched_positions)

# تحميل البيانات
joblist_path = r"C:\Users\salma.elsayed\Desktop\basic data\rah\joblist.xlsx"
joblist_df = pd.read_excel(joblist_path, sheet_name='joblist main')
sap_df = pd.read_excel(joblist_path, sheet_name='sap_')

# تنظيف النصوص (إزالة المسافات الزائدة وتحويل النص إلى أحرف صغيرة)
joblist_df['Position - Job List'] = joblist_df['Position - Job List'].str.strip().str.lower()
sap_df['Position'] = sap_df['Position'].str.strip().str.lower()

# تشغيل عملية المطابقة
matched_df, unmatched_df = find_matching_positions(joblist_df, sap_df, threshold=85)

# دمج matched_df و unmatched_df في DataFrame واحد
final_df = pd.concat([matched_df, unmatched_df], ignore_index=True)
```

```python
from datetime import datetime

#Convert 'Hiring Date' column to datetime format
basic_data['Hiring Date'] = pd.to_datetime(basic_data['Hiring Date'], errors='coerce')

# Function to calculate years and months between the hiring date and today
def calculate_years_months(date):
    today = datetime.today()
    # Calculate years difference
    years = today.year - date.year - ((today.month, today.day) < (date.month, date.day))
    # Calculate months difference after accounting for full years
    months = today.month - date.month - (today.day < date.day)
    if months < 0:
        months += 12
    return f"{years} years, {months} Months"

# Apply the function to create a new 'Years of Service' column
basic_data['Years of Service'] = basic_data['Hiring Date'].apply(calculate_years_months)


# Define a function to categorize Tenurity based on Years of Service
def categorize_tenurity(years_of_service):
    years = int(years_of_service.split()[0])  # Extract the year as an integer
    
    if years < 1:
        return "Probation" if "Months" in years_of_service and int(years_of_service.split()[2]) <= 6 else "7 - 12 Months"
    elif years < 2:
        return "1 - 2 Years"
    elif years < 5:
        return "2 - 5 Years"
    elif years < 10:
        return "5 - 10 Years"
    else:
        return "10+ Years"

# Apply the categorization function
basic_data['Tenurity'] = basic_data['Years of Service'].apply(categorize_tenurity)

# Map Tenurity to sorting values for the "Tenurity Sorting" column
tenurity_order = {
    "Probation": 1,
    "4 - 6 Months": 2,
    "7 - 12 Months": 3,
    "1 - 2 Years": 4,
    "2 - 5 Years": 5,
    "5 - 10 Years":6,
    "10+ Years": 7
}
basic_data['Tenurity Sorting'] = basic_data['Tenurity'].map(tenurity_order)

#basic_data.drop(['Job','ACTION MEMO','Managerial Duties',],inplace=True, axis=1)

```
